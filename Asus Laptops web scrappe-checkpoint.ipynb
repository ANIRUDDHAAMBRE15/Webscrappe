```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
URL = "https://www.flipkart.com/computers/laptops/~asus-gaming-laptops21/pr?sid=6bo%2Cb5g&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIkFzdXMgR2FtaW5nIExhcHRvcHMiXSwidmFsdWVUeXBlIjoiTVVMVElfVkFMVUVEIn19fX19&wid=5.productCard.PMU_V2_3"
```

```python
import requests
r = requests.get(URL)
htmlcontent =r.content
htmlcontent
```

```python
soup = BeautifulSoup(htmlcontent,"html.parser")
```

```python
#Access Asus laptops name by using this class.
```

```python
Laptop = soup.find_all("div",class_ ="_4rR01T")
Laptop
```

```python
#Access the pricing during a sale by using this class.
```

```python
Laptop_Prices = soup.find_all(class_="_30jeq3 _1_WHN1")
Laptop_Prices
```

```python
list_Laptop_Prices=[]

for i in range(0,len(Laptop_Prices)):
    list_Laptop_Prices.append(Laptop_Prices[i].get_text())
    
    #print(Mobile_names[i].get_text())
print(list_Laptop_Prices)
```

```python
list_Laptop=[]
for i in range(0,len(Laptop)):
    list_Laptop.append(Laptop[i].get_text())
    #print(Mobile_names[i].get_text())
print(Laptop)
```

```python
#Then access the before discount prices in detail by using this class.
```

```python
discount = soup.find_all(class_="_3I9_wc _27UcVY")
discount
```

```python
list_discount_Prices=[]

for i in range(0,len(discount)):
    list_discount_Prices.append(discount[i].get_text())
    
    #print(Mobile_names[i].get_text())
print(list_discount_Prices)
```

```python
#Then arrange properly in column all above data by using import pandas library.
```

```python
import pandas as pd
df=pd.DataFrame({"Laptop_Names":list_Laptop,"Discounted Prices":list_Laptop_Prices,"Before sale":list_discount_Prices})
df
```
